Here, we find which values do we want for further data analysis. To check this we run "value_counts()" on the selected columns. First, we run value_counts for CASE_STATUS and VISA_CLASS only and choose only the categories we want to select.

```python
print("Value Counts for 'CASE_STATUS': \n",df['CASE_STATUS'].value_counts())
# we would keep only "CERTIFIED" and "DENIED"

print("\n=========================================================================================\n"
      "Value Counts for 'VISA_CLASS': \n",df['VISA_CLASS'].value_counts())
# Select only "H-1B"
```

We only select 'H-1B' class for VISA_CLASS and 'DENIED' and 'CERTIFIED' cases for CASE_STATUS. This reduces are our rows by nearly 80,000.

```python
df_1 = df[df['VISA_CLASS'].isin(['H-1B'])]
df_2 = df_1[df_1['CASE_STATUS'].isin(['DENIED','CERTIFIED'])]

df_2.shape
```

```python
#in order to reduce categories we will remove this employee specific information

#EMPLOYER_NAME
print("\n=========================================================================================\n"
      "Value Counts for 'EMPLOYER_NAME': \n",df_2['EMPLOYER_NAME'].value_counts())


#EMPLOYER_CITY
print("\n=========================================================================================\n"
      "Value Counts for 'EMPLOYER_CITY':\n",df_2['EMPLOYER_CITY'].value_counts())


#EMPLOYER_STATE
print("\n=========================================================================================\n"
      "Value Counts for 'EMPLOYER_STATE': \n",df_2['EMPLOYER_STATE'].value_counts())
```
